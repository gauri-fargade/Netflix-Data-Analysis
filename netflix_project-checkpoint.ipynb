```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

sns.set(style='whitegrid')
```

```python
data = pd.read_csv("netflix1.csv")
data.head()
```

```python
data.info()
data.shape
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

```python

print("Columns:", data.columns.tolist())


print("\nMissing values:\n")
print(data.isnull().sum())


columns_to_check = ['director', 'cast', 'country']
existing_columns = [col for col in columns_to_check if col in data.columns]


data.dropna(subset=existing_columns, inplace=True)
```

```python

data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')


print(data['date_added'].head())
print("\nData types after conversion:\n")
print(data.dtypes)
```

```python

type_counts = data['type'].value_counts()


plt.figure(figsize=(6, 4))
sns.barplot(x=type_counts.index, y=type_counts.values, palette='Set2')

plt.title('Distribution of Content by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python

data['genres'] = data['listed_in'].apply(lambda x: x.split(', '))


all_genres = sum(data['genres'], [])


genre_counts = pd.Series(all_genres).value_counts().head(10)


plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='Set3')

plt.title('Most Common Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

```python

data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month


plt.figure(figsize=(12, 6))
sns.countplot(x='year_added', data=data, palette='coolwarm')

plt.title('Content Added Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

top_directors = data['director'].dropna().value_counts().head(10)


plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='Blues_d')

plt.title('Top 10 Directors with the Most Titles on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.tight_layout()
plt.show()
```

```python

movie_titles = data[data['type'] == 'Movie']['title'].dropna()


wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='black',
    colormap='Set2'
).generate(' '.join(movie_titles))


plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Netflix Movie Titles')
plt.tight_layout()
plt.show()
```

```python

data['month'] = data['date_added'].dt.month


monthly_movies = data[data['type'] == 'Movie']['month'].value_counts().sort_index()
monthly_shows = data[data['type'] == 'TV Show']['month'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
plt.plot(monthly_movies.index, monthly_movies.values, label='Movies', marker='o')
plt.plot(monthly_shows.index, monthly_shows.values, label='TV Shows', marker='s')

plt.title("Monthly Releases of Movies and TV Shows on Netflix")
plt.xlabel("Month")
plt.ylabel("Number of Releases")
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                                       'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

yearly_movies = data[data['type'] == 'Movie']['year_added'].value_counts().sort_index()
yearly_shows = data[data['type'] == 'TV Show']['year_added'].value_counts().sort_index()


plt.figure(figsize=(12, 6))
plt.plot(yearly_movies.index, yearly_movies.values, label='Movies', marker='o')
plt.plot(yearly_shows.index, yearly_shows.values, label='TV Shows', marker='s')

plt.title("Yearly Releases of Movies and TV Shows on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

top_countries = data['country'].value_counts().head(10)


plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')

plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python

top_movie_genres = data[data['type'] == 'Movie'].groupby('listed_in').size().sort_values(ascending=False).head(10)
top_tv_genres = data[data['type'] == 'TV Show'].groupby('listed_in').size().sort_values(ascending=False).head(10)


plt.figure(figsize=(10, 5))
plt.barh(top_movie_genres.index, top_movie_genres.values, color='salmon')
plt.title("Top 10 Movie Genres on Netflix")
plt.xlabel("Number of Movies")
plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 5))
plt.barh(top_tv_genres.index, top_tv_genres.values, color='skyblue')
plt.title("Top 10 TV Show Genres on Netflix")
plt.xlabel("Number of TV Shows")
plt.tight_layout()
plt.show()
```

```python

top_15_directors = data['director'].value_counts().dropna().head(15)


plt.figure(figsize=(12, 6))
sns.barplot(x=top_15_directors.values, y=top_15_directors.index, palette='magma')

plt.title("Top 15 Directors with Most Netflix Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```
